```python
from transformers import BertModel,BertTokenizer
from bertviz import head_view
import pandas as pd
import torch
```

```python
text="my name is chandan , i am in IITH for Workshop"
```

```python
token_bert=BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
token=token_bert.encode(text)
```

```python
#token
```

```python
inputs=torch.tensor(token).unsqueeze(0)
```

```python
inputs
```

```python
token_as_list=token_bert.convert_ids_to_tokens(inputs[0])
```

```python
#token_as_list
```

```python
model=BertModel.from_pretrained('bert-base-uncased')
```

```python
attentions=model(inputs,output_attentions=True)[2]
```

```python
#attions
```

```python
head_view(attentions,token_as_list)
```
